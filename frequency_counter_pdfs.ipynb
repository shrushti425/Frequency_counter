```python
from pypdf import PdfReader

reader = PdfReader("assignment.pdf")
s=""
print(len(reader.pages))
for page in reader.pages:
  s+=page.extract_text()
s1=s.lower()
l=s1.split()

d={}
for i in l:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
print("Frequency of each word in the PDF:")
print(d)
z=list(d.values())
maxi=max(z)
print('\nWord with maximum frequency:')
for key,value in d.items():
    if value==maxi:
        print(key)
```
